#### Plotting Graphs

```python
corr = data.corr()['Class'].drop('Class').sort_values()
corr.plot(kind='bar', figsize=(15,5))
plt.title('Feature Correlation with Fraud')
plt.show()
```

```python
sns.histplot(data['Amount'], bins=50, kde=True)
plt.title('Transaction Amount Distribution')
plt.show()
```

```python
sns.kdeplot(data[data['Class'] == 1]['Amount'], label = 'Fraud', shade = True)
sns.kdeplot(data[data['Class'] == 0]['Amount'], label = 'Legit', shade = True)
plt.title('Amount Distribution: Fraud v/s Legit')
plt.legend()
plt.show()
```

```python
sns.kdeplot(data[data['Class'] == 1]['Time'], label = 'Fraud', shade = True)
sns.kdeplot(data[data['Class'] == 0]['Time'], label = 'Legit', shade = True)
plt.title('Time Distribution: Fraud v/s Legit')
plt.legend()
plt.show()
```
